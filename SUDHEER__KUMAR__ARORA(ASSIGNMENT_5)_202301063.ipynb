```python
import pandas as pd
import io
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
comp=pd.read_csv("company_sales_data (1).csv")
comp
```

```python
sns.lineplot(x=comp["month_number"],y=comp["total_profit"])
plt.show()
```

```python
sns.lineplot(x=comp["month_number"],y=comp["facecream"],label="facecream")
sns.lineplot(x=comp["month_number"],y=comp["facewash"],label="facewash")
sns.lineplot(x=comp["month_number"],y=comp["toothpaste"],label="toothpaste")
sns.lineplot(x=comp["month_number"],y=comp["bathingsoap"],label="bathingsoap")
sns.lineplot(x=comp["month_number"],y=comp["shampoo"],label="shampoo")
sns.lineplot(x=comp["month_number"],y=comp["moisturizer"],label="moiturizer")
plt.legend()
plt.show()
```

```python
sns.barplot(x=comp["month_number"],y=comp["facecream"],label="facecream")
sns.barplot(x=comp["month_number"],y=comp["facewash"],label="facewash")
sns.barplot(x=comp["month_number"],y=comp["toothpaste"],label="toothpaste")
sns.barplot(x=comp["month_number"],y=comp["bathingsoap"],label="bathingsoap")
sns.barplot(x=comp["month_number"],y=comp["shampoo"],label="shampoo")
sns.barplot(x=comp["month_number"],y=comp["moisturizer"],label="moiturizer")
plt.xlabel('month')
plt.ylabel('sales')
plt.title("Bar Graph")
plt.legend()
plt.show()
```
