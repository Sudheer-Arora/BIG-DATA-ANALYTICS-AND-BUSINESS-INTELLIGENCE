```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("company_sales_data (1).csv")
data
```

```python
data.info()
```

```python
data.head(10)
```

```python
plt.bar(data["month_number"],data["facecream"],label="facecream")
plt.bar(data["month_number"],data["facewash"],label="facewash")
plt.xlabel('month')
plt.ylabel('sales')
plt.title("Bar chart")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
data['total_profit'].plot(kind='hist',bins=5,title='Total Profit Histogram')
plt.xlabel('Total Profit Range')
plt.ylabel('Frequency')
plt.show()
```
