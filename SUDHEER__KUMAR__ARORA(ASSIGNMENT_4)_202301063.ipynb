```python
import pandas as pd
import io
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
comp=pd.read_csv("company_sales_data (1).csv")
comp
```

```python
plt.plot(comp["month_number"],comp["total_profit"],color='green',marker='+',markersize =12)
plt.title("Line plot")
plt.xlabel('Month no.')
plt.ylabel('Total profit')
plt.show()
```

```python
plt.plot(comp["month_number"],comp["facecream"],label="facecream")
plt.plot(comp["month_number"],comp["facewash"],label="facewash")
plt.plot(comp["month_number"],comp["toothpaste"],label="toothpaste")
plt.plot(comp["month_number"],comp["bathingsoap"],label="bathingsoap")
plt.plot(comp["month_number"],comp["shampoo"],label="shampoo")
plt.plot(comp["month_number"],comp["moisturizer"],label="moiturizer")
plt.title("multi Line plot")
plt.xlabel('month')
plt.ylabel('sales')
plt.legend()
plt.show()
```

```python
plt.bar(comp["month_number"],comp["facecream"],label="facecream")
plt.bar(comp["month_number"],comp["facewash"],label="facewash")
plt.bar(comp["month_number"],comp["toothpaste"],label="toothpaste")
plt.bar(comp["month_number"],comp["bathingsoap"],label="bathingsoap")
plt.bar(comp["month_number"],comp["shampoo"],label="shampoo")
plt.bar(comp["month_number"],comp["moisturizer"],label="moiturizer")
plt.xlabel('month')
plt.ylabel('sales')
plt.title("Bar Graph")
plt.legend()
plt.show()
```

```python
plt.scatter(comp["month_number"],comp["facecream"],label="facecream")
plt.scatter(comp["month_number"],comp["facewash"],label="facewash")
plt.scatter(comp["month_number"],comp["toothpaste"],label="toothpaste")
plt.title("scatter plot")
plt.xlabel('month')
plt.ylabel('sales')
plt.legend()
plt.show()
```
