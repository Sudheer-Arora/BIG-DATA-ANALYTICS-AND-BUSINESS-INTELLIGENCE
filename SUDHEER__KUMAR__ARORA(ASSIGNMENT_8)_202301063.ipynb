```python
import numpy as np
import pandas as pd
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from mpl_toolkits.mplot3d import Axes3D
```

```python
iris=sklearn.datasets.load_iris()
X=iris.data[:,2:]
Y=iris.target
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.2,random_state=21)
```

```python
model=LinearRegression()
model.fit(Xtrain,Ytrain)
```

```python
accuracy=model.score(Xtest,Ytest)
accuracy
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
f1=Xtrain[:,0]
f2=Xtrain[:,1]
ax.scatter(f1,f2,Ytrain,color='blue')
X_f1=Xtrain[:,0]
X_f2=Xtrain[:, 0]
x1_range=np.linspace(X_f1.min(),X_f1.max(),100)
x2_range=np.linspace(X_f2.min(),X_f2.max(),100)
x1_mesh,x2_mesh=np.meshgrid(x1_range, x2_range)
predicted_y=model.predict(np.c_[x1_mesh.ravel(),x2_mesh.ravel()]).reshape(x1_mesh.shape)
ax.plot_surface(x1_mesh,x2_mesh,predicted_y,color='red',alpha=0.5)
ax.set_xlabel('F 1')
ax.set_ylabel('F 2')
ax.set_title('Multiple Linear Regression')
plt.show()
```
